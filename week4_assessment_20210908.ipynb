```python
dic = {'X2 house age':'x2_age','X3 distance to the nearest MRT station':'x3_mtr',
       'X4 number of convenience stores':'x4_stores','X5 latitude':'x5_lat','X6 longitude':'x6_long',
       'Y house price of unit area':'y_price'}
df1=df.rename(columns=dic)
df1.head
```

```python
x=df1[['x2_age','x3_mtr','x4_stores','x5_lat','x6_long']]
x
```

```python
#Question 6
def split():
    return train_test_split(df1[['x2_age','x3_mtr','x4_stores','x5_lat','x6_long']],df1['y_price'],
                            test_size=0.2, random_state=42)
    
X_train, X_test, y_train, y_test = split()
X_train
```

```python
#Question 9
def makeModel():
    # Create linear regression object
    model=linear_model.LinearRegression()
    # Train the model using the training sets
    model.fit(X_train,y_train)
    # Make predictions using the testing set
    y_pred = model.predict(X_test)
    #fill in the blanks, how to get the value of coefficients, intercept, mse, rmse, r2score
    coefficients = model.coef_
    intercept = model.intercept_
    mse = mean_squared_error(y_test, y_pred)
    rmse = math.sqrt(mse)
    r2score = r2_score(y_test, y_pred)
#SOLUTION END


    return (coefficients,intercept, mse,rmse, r2score)

makeModel()
```
